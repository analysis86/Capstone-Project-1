#  Calculate age at sale

```python
df1['AgeAtSale'] = df1['SaleYr'] - df1['YearBuilt']
```

```python
df1['YearsSinceRemodel'] = df1['SaleYr'] - df1['YearRemodAdd']
```

#  Display the first few rows to verify the new columns

```python
print(df1[['AgeAtSale', 'YearsSinceRemodel']].head())
```

#  Calculate Total Bathrooms

```python
df1['TotalBathrooms'] = df1['BsmtFullBath'] + df1['BsmtHalfBath']*0.5 + df1['Bath1'] + df1['Bath2']*0.5
```

```python
print(df1[['TotalBathrooms']].head())
```

# Calculate total square foot

```python
df1['TotalSqFootage'] = df1['1stFlrSF'] + df1['2ndFlrSF'] + df1['BsmtFinSF2'] + df1['BsmtUnfSF']
```

```python
print(df1[['TotalSqFootage']].head())
```

# Calculate Quality Score:

```python
df1['QualityScore'] = df1['OverallQual'] * df1['OverallCond']
```

```python
print(df1[['QualityScore']].head())
```

# Calculate Interaction Features:

```python
df1['GrLivArea_OverallQual'] = df1['GrLivArea'] * df1['OverallQual']
```
